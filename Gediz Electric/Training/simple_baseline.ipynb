```python
df_train=df_train.set_index('tarih')
df_test=df_test.set_index('tarih')
```

```python
df_train.index=pd.to_datetime(df_train.index)
df_test.index=pd.to_datetime(df_test.index)
```

```python
def plot_target(df,ilce_0,ilce_1):
    temp=df[(df['ilce_0']==ilce_0) & (df['ilce_1']==ilce_1)]
    temp['bildirimsiz_sum'].plot(style='.',figsize=(15,5),title='Power Outages',color='blue')
```

```python
train=df_train.loc[df_train.index < '2023-06-01']
val=df_train.loc[df_train.index >= '2023-06-01']
```

```python
fig , ax = plt.subplots(figsize=(15,5))
train['bildirimsiz_sum'].plot(ax=ax,label='Training_Set')
val['bildirimsiz_sum'].plot(ax=ax,label='Val_Set')
plt.show()
```

# Feature Creation

```python
def create_features(df):
    """
    Create Time Series Features based on index datetime
    """
    df['day_of_week']=df.index.dayofweek 
    df['month']=df.index.month
    df['year']=df.index.year
    df['quarter']=df.index.quarter
    df['day_of_year']=df.index.dayofyear

    return df
```

```python
train=create_features(train)
val=create_features(val)
df_test=create_features(df_test)
```

# Modeling

```python
train.columns
FEATURES=['ilce_0', 'ilce_1', 'bildirimli_sum', 't_2m:C_min',
       't_2m:C_max', 't_2m:C_mean', 'effective_cloud_cover:p_min',
       'effective_cloud_cover:p_max', 'effective_cloud_cover:p_mean',
       'global_rad:W_min', 'global_rad:W_max', 'global_rad:W_mean',
       'relative_humidity_2m:p_min', 'relative_humidity_2m:p_max',
       'relative_humidity_2m:p_mean', 'wind_dir_10m:d_min',
       'wind_dir_10m:d_max', 'wind_dir_10m:d_mean', 'wind_speed_10m:ms_min',
       'wind_speed_10m:ms_max', 'wind_speed_10m:ms_mean',
       'prob_precip_1h:p_min', 'prob_precip_1h:p_max', 'prob_precip_1h:p_mean',
       't_apparent:C_min', 't_apparent:C_max', 't_apparent:C_mean',
       'Bayram_Flag', 'day_of_week', 'month', 'year', 'quarter',
       'day_of_year']

TARGET=['bildirimsiz_sum']
```

```python
X_train=train[FEATURES]
y_train=train[TARGET]

X_val=val[FEATURES]
y_val=val[TARGET]
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
from xgboost import XGBRegressor


reg=XGBRegressor(n_estimators=2000,
                early_stopping_rounds=50,learning_rate=0.01,col_sample=0.8)

reg.fit(X_train,y_train,
        eval_set=[(X_train,y_train),(X_val,y_val)],
        eval_metric='mae',
        verbose=50)
```

# Feature_Importances

```python
fi=pd.DataFrame(data=reg.feature_importances_,index=reg.feature_names_in_,
columns=['importances'])
```

```python
fi.sort_values('importances').plot(kind='barh',title='feature_importances')
```

```python
df_train
```

```python
val['prediction']=reg.predict(X_val)
df_train=df_train.merge(val[['prediction']],how='left',left_index=True,right_index=True)
ax=df_train[['bildirimsiz_sum']].plot(figsize=(15,5))
df_train['prediction'].plot(ax=ax,style='.')
plt.show()
```

# Evaluation
