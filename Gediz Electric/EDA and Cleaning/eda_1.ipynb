```python
"""
Cell generated by Data Wrangler.
"""
import pandas as pd

def clean_data(train):
    # Split text using string '-' in column: 'ilce'
    loc_0 = train.columns.get_loc('ilce')
    train_split = train['ilce'].str.split(pat='-', expand=True).add_prefix('ilce_')
    train = pd.concat([train.iloc[:, :loc_0], train_split, train.iloc[:, loc_0:]], axis=1)
    train = train.drop(columns=['ilce'])
    return train

train_clean = clean_data(train.copy())
train_clean.head()
```

```python
clean_weather.date=pd.to_datetime(clean_weather.date)
```

```python
clean_weather
```

```python
clean_weather['date']=clean_weather.date.dt.date
```

```python
grouped=clean_weather.groupby(['date', 'ilce_0', 'ilce_1'])
```

```python
aggregated_weather = grouped.agg({
    't_2m:C': ['min', 'max', 'mean'],
    'effective_cloud_cover:p': ['min', 'max', 'mean'],
    'global_rad:W': ['min', 'max', 'mean'],
    'relative_humidity_2m:p': ['min', 'max', 'mean'],
    'wind_dir_10m:d': ['min', 'max', 'mean'],
    'wind_speed_10m:ms': ['min', 'max', 'mean'],
    'prob_precip_1h:p': ['min', 'max', 'mean'],
    't_apparent:C': ['min', 'max', 'mean']
})
```

```python
aggregated_weather.columns = ['_'.join(col).strip() for col in aggregated_weather.columns.values]

# Reset index to make 'date', 'ilce_0', and 'ilce_1' regular columns again
aggregated_weather.reset_index(inplace=True)
```
