```python
import pandas as pd
import numpy as np 
train=pd.read_csv(r'C:\Users\enesm\OneDrive\Masaüstü\KAGGLE\Iyzico\Data\2-Sorted and Equalized\train_sorted.csv')
test=pd.read_csv(r'C:\Users\enesm\OneDrive\Masaüstü\KAGGLE\Iyzico\Data\2-Sorted and Equalized\test_sorted.csv')
```

```python
pd.options.display.max_columns=None
```

```python
train['is_train']=True
test['is_train']=False
```

```python
tt=pd.concat([train,test],axis=0)
```

```python
tt  
```

# Onehot

```python
tt
```

```python
['mcc_id','settlement_period','working_type','merchant_segment']
```

```python
for col in ['merchant_source_name','settlement_period','working_type','merchant_segment']:
    temp_dummies=pd.get_dummies(tt[col],dtype=int)
    tt=pd.concat([tt,temp_dummies],axis=1)
    tt=tt.drop(col,axis=1)
```

```python
for i in tt.columns:
    print(i)
```

# Label Encode

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for col in ['mcc_id']:
    tt[col]=le.fit_transform(tt[col])
```

```python
train=tt[tt['is_train']==1]
test=tt[tt['is_train']==0]
```

```python
train.drop('is_train',axis=1,inplace=True)
test.drop('is_train',axis=1,inplace=True)
```

```python
train.to_csv('train_encoded.csv',index=False)
```
