```python
import pandas as pd

# Sample data
# Assuming train and test_sorted DataFrames are already defined

# Selecting relevant columns from the train set for merging
train_features = train[['merchant_id', 'merchant_source_name', 'settlement_period', 'working_type', 'mcc_id', 'merchant_segment']]

# Aggregating features in case of multiple entries for the same merchant_id in the train set
# Here, using the mode (most frequent value) for each feature, but you can choose other methods like mean or median for numerical features
aggregated_features = train_features.groupby('merchant_id').agg(lambda x: pd.Series.mode(x).iloc[0])

# Merging the aggregated features into the test set
test_enriched = test_sorted.merge(aggregated_features, on='merchant_id', how='left')

print(test_enriched)
```
