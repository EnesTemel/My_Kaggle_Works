# Categorizing same merchants

```python
cols_to_merge = ['mcc_id','Merchant Source - 1', 'Merchant Source - 2', 'Merchant Source - 3', 
                 'Settlement Period - 1', 'Settlement Period - 2', 'Settlement Period - 3',
                 'Working Type - 1', 'Working Type - 2', 'Working Type - 3', 'Working Type - 4', 
                 'Working Type - 5', 'Working Type - 6', 'Segment - 1', 'Segment - 2', 
                 'Segment - 3', 'Segment - 4']
```

```python
tt['category']=tt[cols_to_merge].astype(str).agg(''.join,axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
tt.category=le.fit_transform(tt['category'])
```
