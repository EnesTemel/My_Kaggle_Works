```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
import seaborn as sns 
```

```python
test=pd.read_csv(r'C:\Users\enesm\OneDrive\Masaüstü\KAGGLE\Anadolu Hayat Emeklilik\Data\Encoded\test_encoded.csv')
train=pd.read_csv(r'C:\Users\enesm\OneDrive\Masaüstü\KAGGLE\Anadolu Hayat Emeklilik\Data\Encoded\train_encoded.csv')
```

```python
train['is_train']=1
test['is_train']=0
```

```python
all_data=pd.concat([train,test],axis=0)
all_data
```

```python
for column in all_data.columns:
    # Check if the column is numerical and not the 'FLAG' column
    if all_data[column].dtype in ['int64', 'float64'] and column != 'FLAG':
        plt.figure(figsize=(10, 6))
        sns.boxplot(x='is_train', y=column, data=all_data)
        plt.title(f'Distribution of {column} Across Months')
        plt.xlabel('Month')
        plt.ylabel(column)
        plt.show()
```
